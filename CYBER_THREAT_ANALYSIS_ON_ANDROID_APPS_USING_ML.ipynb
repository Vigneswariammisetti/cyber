```python
import pandas as pd
import numpy as np
import random

# Parameters
num_records = 1000

# Generate synthetic data
data = []
for _ in range(num_records):
    permissions = random.randint(5, 50)  # Number of permissions
    activities = random.randint(1, 20)  # Number of activities
    services = random.randint(1, 10)  # Number of services
    malicious = random.choice([0, 1])  # 0: Benign, 1: Malicious

    data.append([permissions, activities, services, malicious])

# Create DataFrame
columns = ['Permissions', 'Activities', 'Services', 'Malicious']
df = pd.DataFrame(data, columns=columns)

# Display sample data
print(df.head())

# Save to CSV
df.to_csv('synthetic_android_app_dataset.csv', index=False)
```

```python
# Load the synthetic dataset
df = pd.read_csv('synthetic_android_app_dataset.csv')

# Display the preprocessed data
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Feature and target variables
X = df[['Permissions', 'Activities', 'Services']]
y = df['Malicious']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
# Train the Random Forest model6
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)

# Predict on the test set
y_pred = clf.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f'Accuracy: {accuracy}')
print('Classification Report:')
print(report)
```

```python
# Function to predict app safety
def predict_app_safety(features):
    prediction = clf.predict([features])
    return "Malicious" if prediction[0] == 1 else "Benign"

# Example usage with new sample features
new_sample_features = [30, 10, 5]  # Example values for permissions, activities, services
app_safety = predict_app_safety(new_sample_features)
print(f'The app is: {app_safety}')
```
